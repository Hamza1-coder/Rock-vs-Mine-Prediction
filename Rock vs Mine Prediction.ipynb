```python
# importing the dependencies
import numpy as np
import pandas as pd
from matplotlib import pyplot
```

```python
# loading the dataset
df = pd.read_csv('sonar.csv', header=None)
df
```

```python
df.describe()
```

```python
# Histogram 
df.hist(sharex=False, sharey=False, xlabelsize=1, ylabelsize=1, figsize = (12,12))
pyplot.show()
```

```python
# density
df.plot(kind='density', subplots =True,layout =(8,8), sharex=False, legend=False, figsize=(12,12))
pyplot.show()
```

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
# M represent the mine
# R represent the rock
# we see there are a clear differnce in the means values of Mine and Rock
```

```python
X = df.drop(columns=60, axis=1)
y = df[60]
print("Shape of X: " ,X.shape)
print("Shape of y: ", y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.1, random_state= 1)
```

```python
from sklearn.linear_model import LogisticRegression
lgr = LogisticRegression()
```

```python
lgr.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy Score train data: ",accuracy_score(lgr.predict(X_train), y_train))
```

```python
print("Accuracy Score test data: ",accuracy_score(lgr.predict(X_test), y_test))
```

## Tester Function

```python
def prediction(lst):
    lst = np.array(lst)
    _list = lst.reshape(1,-1)
    predict = lgr.predict(_list)
    if predict[0] == 'R':
        print("Object is Rock")
    else:
        print("Object is Mine")
```

#### Testing Model for both output
